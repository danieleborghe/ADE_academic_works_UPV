## Preliminary operations

```python
# import main libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
from factor_analyzer import FactorAnalyzer
import seaborn as sns
from sklearn.cluster import KMeans
sns.set(style="white")
```

```python
# import reflex C dataset
df_reflex_C = pd.read_excel("reflex_c.xlsx")
# get job values features
df_job_values = df_reflex_C.loc[:, "Work_Autonomy":]
# convert job values to np array for clustering
np_job_values = np.array(df_job_values)

df_job_values
```

## Factor transformation

```python
# fit factor analysis model with the 3 factors obtained
n_factors = 3
factor_analyzer = FactorAnalyzer(n_factors, rotation=None)
factor_analyzer.fit(df_job_values)

# calculate factor scores
factor_scores = factor_analyzer.transform(df_job_values)
factor_scores
```

## Hierarchical clustering

### Reduced dataset

```python
# calculate hierarchical clustering matrix with Ward method
linkage_matrix = linkage(factor_scores, method='ward', metric='euclidean')

# plot hierachical clustering dendrogram
plt.figure(figsize=(14, 5))

dendrogram(linkage_matrix)

plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram (Ward linkage)')

plt.savefig("hierarchical_factors_ward.pdf", format="pdf")

plt.show()
```

```python
# calculate hierarchical clustering matrix with single linkage
linkage_matrix = linkage(factor_scores, method='single', metric='euclidean')

# plot hierachical clustering dendrogram
plt.figure(figsize=(14, 5))

dendrogram(linkage_matrix)

plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram (single linkage)')

plt.savefig("hierarchical_factors_single.pdf", format="pdf")

plt.show()
```

```python
# calculate hierarchical clustering matrix with complete linkage
linkage_matrix = linkage(factor_scores, method='complete', metric='euclidean')

# plot hierachical clustering dendrogram
plt.figure(figsize=(14, 5))

dendrogram(linkage_matrix)

plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram (complete linkage)')

plt.savefig("hierarchical_factors_complete.pdf", format="pdf")

plt.show()
```

```python
# calculate hierarchical clustering matrix with average linkage
linkage_matrix = linkage(factor_scores, method='average', metric='euclidean')

# plot hierachical clustering dendrogram
plt.figure(figsize=(14, 5))

dendrogram(linkage_matrix)

plt.xlabel('Data Points')
plt.ylabel('Distance')
plt.title('Hierarchical Clustering Dendrogram (average linkage)')

plt.savefig("hierarchical_factors_average.pdf", format="pdf")

plt.show()
```

## K-Means

```python
# set the number of clusters
n_clusters = 4

# create K-Means instance
kmeans = KMeans(n_clusters = n_clusters, n_init="auto")

# fit K-Means
kmeans.fit(factor_scores)

# get clustering labels
labels = kmeans.labels_
# get clustering centers
cluster_centers = kmeans.cluster_centers_

# plot the clustering scatterplot
plt.figure(figsize=(14, 5))

sns.scatterplot(x=factor_scores[:, 0], y=factor_scores[:, 1], hue=labels, palette='viridis')
sns.scatterplot(x=cluster_centers[:, 0], y=cluster_centers[:, 1], color='red', marker='X', s=100, label='Cluster Centers')

plt.xlabel('Factor 1')
plt.ylabel('Factor 2')
plt.title('K-Means Clustering (factors)')
plt.legend()

plt.savefig("kmeans_factors_4k.pdf", format="pdf")

plt.show()
```
